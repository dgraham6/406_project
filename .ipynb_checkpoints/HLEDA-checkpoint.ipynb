```python
import pandas as pd
import matplotlib.pyplot as plt

# Identify volatility columns
v_cols = [c for c in df.columns if c.startswith("V")]

# Compute std dev to find most volatile features
volatility = df[v_cols].std().sort_values(ascending=False)

# Select top 5 most volatile
top_vol_cols = volatility.head(5).index.tolist()

# Plot only the most volatile features
plt.figure(figsize=(12, 6))

for col in top_vol_cols:
    plt.plot(df["date_id"], df[col], label=col)

plt.xlabel("date_id")
plt.ylabel("Volatility feature value")
plt.title("Top 5 Most Volatile Volatility Features Over Time")
plt.legend()
plt.tight_layout()
plt.show()
```
